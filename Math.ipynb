```python
def countAsterisks(s: str):
    flag=True
    count=0
    for i in s:
        if i=="|":
            flag=not(flag)
        if flag and i=="*":
            count+=1
    return count
```

```python
def reverse(x):
    sign=1 if x>0 else -1
    x=abs(x)
    ans=0
    while x>0:
        temp=x%10
        ans=ans*10+temp
        x//=10
    ans=ans*sign
    if ans>2**31-1 or ans<-2**31:
        return 0
    return ans
```

```python
def removeDuplicates(self, nums: List[int]) -> int:
    p=0
    for i in range(1,len(nums)):
        if nums[i]!=nums[p]:
            p+=1
            nums[p]=nums[i]
    return p+1
```

```python
class Solution:
    def maxArea(self, height: List[int]) -> int:
        left=0
        right=len(height)-1
        ans=0
        while left<right:
            if height[left]>height[right]:
                temp=(right-left)*height[right]
                right-=1
            else:
                temp=(right-left)*height[left]
                left+=1
            ans=max(ans,temp)
        return ans

        
```

```python
class Solution:
    def isPalindrome(self, s: str) -> bool:
        l=0
        r=len(s)-1
        s=s.lower()
        while l<r:
            if not s[l].isalnum():
                l+=1
                continue
            if not s[r].isalnum():
                r-=1
                continue
            if s[l]!=s[r]:
                #print(s[l],s[r])
                return False
            l+=1
            r-=1
        return True


```

```python
class Solution:
    def judgeSquareSum(self, c: int) -> bool:
        a=0
        b=int(c**0.5)+1
        while a<b:
            temp=a*a+b*b
            if temp==c:
                return True
            if temp>c:
                b-=1
            else:
                a+=1
        return False
```

```python
class Solution:
    def moveZeroes(self, nums: List[int]) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        p=0
        for i in range(len(nums)):
            if nums[i]!=0:
                nums[p],nums[i]=nums[i],nums[p]
                p+=1
        
        
```

```python

class Solution:
    def mergeAlternately(self, word1: str, word2: str) -> str:
        i=0
        m=len(word1)
        n=len(word2)
        ans=""
        while i<m or i<n:
            if i<m:
                ans+=word1[i]
            if i<n:
                ans+=word2[i]
            i+=1
        return ans

```

```python

def isAnagram(self, s: str, t: str) -> bool:
    if len(s)!=len(t):
        return False

    d1={}
    d2={}
    for i in range(len(s)):
        if s[i] in d1:
            d1[s[i]]+=1
        else:
            d1[s[i]]=1

        if t[i] in d2:
            d2[t[i]]+=1
        else:
            d2[t[i]]=1
        
    return d1==d2
        
```
