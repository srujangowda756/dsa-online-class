```python
n=int(input("Enter a number: "))
count=0
while n>0:
    count+=1
    n//=10
    #print(n)
print('----------------')
print(count)
```

```python
def countAsterisks(s: str):
    flag=True
    count=0
    for i in s:
        if i=="|":
            flag=not(flag)
        if flag and i=="*":
            count+=1
    return count
```

```python
def reverse(x):
    sign=1 if x>0 else -1
    x=abs(x)
    ans=0
    while x>0:
        temp=x%10
        ans=ans*10+temp
        x//=10
    ans=ans*sign
    if ans>2**31-1 or ans<-2**31:
        return 0
    return ans
```
